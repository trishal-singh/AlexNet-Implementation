```python
import tensorflow as tf 
import pandas  as pd
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test)=tf.keras.datasets.cifar10.load_data()
```

```python
def process_images(image, label):
    # Normalize images to have a mean of 0 and standard deviation of 1
    #image = tf.image.per_image_standardization(image)
    # Resize images from 32x32 to 70x70
    image = tf.image.resize(image, (70,70))
    return image, label
x_train,y_train=process_images(x_train,y_train)
x_test,y_test=process_images(x_test,y_test)
x_train.shape
```

```python
def Alex_net():
    return tf.keras.models.Sequential([
        
        tf.keras.layers.Conv2D(filters=96, kernel_size=11, strides=4,
                               activation='relu',input_shape=(70,70,3)),
        tf.keras.layers.MaxPool2D(pool_size=3, strides=2),
        
        tf.keras.layers.Conv2D(filters=256, kernel_size=5, padding='same',
                               activation='relu'),
        tf.keras.layers.MaxPool2D(pool_size=3, strides=2),
       
        tf.keras.layers.Conv2D(filters=384, kernel_size=3, padding='same',
                               activation='relu'),
        tf.keras.layers.Conv2D(filters=384, kernel_size=3, padding='same',
                               activation='relu'),
        tf.keras.layers.Conv2D(filters=256, kernel_size=3, padding='same',
                               activation='relu'),
        tf.keras.layers.MaxPool2D(pool_size=3, strides=2),
        tf.keras.layers.Flatten(),
        
        tf.keras.layers.Dense(4096, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(4096, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        # Output layer. Since we are using Fashion-MNIST, the number of
        # classes is 10, instead of 1000 as in the paper
        tf.keras.layers.Dense(100,activation='softmax')
    ])
```

```python
model=Alex_net()
model.compile(loss='sparse_categorical_crossentropy', optimizer=tf.optimizers.SGD(lr=0.001), metrics=['accuracy'])
```

```python
model.build()
model.fit(x=x_train,y=y_train,epochs=50,batch_size=48)
```

```python
model.evaluate(x_test,y_test)
```
